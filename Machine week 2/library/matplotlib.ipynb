# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Data
x = np.arange(15,30,0.1)
y =np.tan(x)
plt.plot(x,y)
plt.title("Sin Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
x = np.arange(15,30,0.1)
y =np.cos(x)
plt.plot(x,y)
plt.title("cos Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
x = np.random.rand(75)
y= np.random.rand(75)
colors = np.random.rand(75)
sizes = np.random.randint(75,150,75)

plt.scatter(x,y, c=colors, s=sizes)
plt.title("Point Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show
```

```python
data = np.random.randn(500)
plt.hist(data, bins=50)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
x = ['A','M','M','A','R']
y = np.random.randint(1,10,5)
plt.bar(x,y)
plt.title("Column Chart")
plt.xlabel("Letters")
plt.ylabel("Numbers")
plt.show()
```

```python
sizes = [50,25,5,2,4,7]
plt.pie(sizes)
plt.title("Pie chart")
plt.show()
```

```python
x = np.arange(-10, 5, 0.5)
y = np.arange(-10, 5, 0.5)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z)

ax.set_title("3D Graphics")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.show()
```
